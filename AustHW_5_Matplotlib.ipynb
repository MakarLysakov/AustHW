```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0,2.1,0.1)


fig = plt.figure()

fig.add_axes([0.11, 0.11, 0.35, 0.8])
plt.plot(a, a*a, 'red')
plt.legend(['$x^2$'])
plt.title('first plot')
plt.ylabel('$y = x^2$')
plt.xlabel('x')
plt.xticks([0,1,2])

fig.add_axes([0.6, 0.11, 0.35, 0.8])
plt.plot(a, a)
plt.legend('x')
plt.title('second plot')
plt.ylabel('y = x')
plt.xlabel('x')
plt.xticks([0,1,2])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math


mas = np.arange(-2, 2.1, 0.1)

fig = plt.figure()

plt.subplot(2,2,1)
plt.plot(mas, np.sin(4*mas), 'red')
plt.xticks([])
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['sin(4x)'], loc = 'center right')
plt.tick_params(axis='both', which='major', labelsize=16)

plt.subplot(2,2,2)
plt.plot(mas, np.cos(mas)*np.sin(mas), 'green')
plt.xticks([])
plt.xlabel('x')
plt.legend(['cos(x)sin(x)'], loc = 'center')
plt.tick_params(axis='both', which='major', labelsize=16)

plt.subplot(2,2,3)
plt.plot(mas, mas**4)
plt.xticks([-2,0,2])
plt.yticks([0,10])
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='both', which='major', labelsize=16)
plt.legend(['x^4'], loc = 'upper center')

plt.subplot(2,2,4)
plt.plot(mas, np.cos(mas) + np.sin(mas), 'black')
plt.xticks([-2,0,2])
plt.xlabel('x')
plt.legend(['cos(x)+sin(x)'], loc = 'lower right')
plt.tick_params(axis='both', which='major', labelsize=16)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f (x,y):
    if x > 0:
        return (1 - y, 1 + x) 
    else:
        return ((-1 + y), (-1 - x))

print(f(0,0))
fig = plt.figure()

plt.xlim([-1, 1])
plt.ylim([-2, 1])
plt.quiver(f(0,0)[0], f(0,0)[1], 1, 1, scale = 5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1,1,20)
y = np.linspace(-2, 1,30)
xx, yy = np.meshgrid(x, y)

def F(x, y):
    return np.where(x > 0, np.array([1 - y, 1 + x]), np.array([-1 + y, -1 - x]))

plt.figure()
plt.xlabel("x")
plt.ylabel("y")
plt.quiver(x, y, F(xx, yy)[0], F(xx, yy)[1], scale=80, label="F(x,y)")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

phi = np.linspace(0, 2*np.pi,20)
r = np.linspace(0, 1, 20)
rr, pp = np.meshgrid(r, phi)
colors = cm.rainbow(rr.flatten())


plt.figure()
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.scatter(rr * np.cos(pp), rr * np.sin(pp), label="circles", color=colors)
plt.legend()
plt.show()

# ellipses
plt.figure()
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.scatter(rr * np.cos(pp), rr * np.sin(pp) * 1/2, label="ellipses", color=colors)
plt.legend()
plt.show()
```
